```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
ts1 = []
ts2 = []
ts3 = []
ts_len = 50
for i in range(ts_len):
    rand_num = np.random.randint(60, 100)
    ts1.append(rand_num)
    ts2.append(rand_num + np.random.randn() * 5)
```

```python
idx_start = 100
idx_end = 250
x_cut = np.arange(idx_start, idx_end)
```

```python
#interplot for smoother line
x = list(range(ts_len))
x_new = np.arange(0,ts_len-1, 0.1)
func = interpolate.interp1d(x,ts1,kind='cubic')
ts1_new = func(x_new)
func = interpolate.interp1d(x,ts2,kind='cubic')
ts2_new = func(x_new)
```

```python

x_new = list(range(len(ts1_new)))
plt.subplot(2,1,1)
plt.plot(x_new, ts1_new, color='lightblue')
plt.plot(x_cut ,ts1_new[idx_start: idx_end], color='steelblue')
plt.axis([-10, 400, 50, 115])
plt.xticks([])
plt.yticks([])


plt.subplot(2,1,2)
plt.plot(x_new, ts2_new, color='bisque')
plt.plot(x_cut ,ts2_new[idx_start: idx_end], color='darkorange')
plt.axis([-10, 400, 50, 115])
plt.xticks([])
plt.yticks([])

plt.savefig('line.png', dpi=500)
```

```python
ts = []
ts_len = 30
for i in range(ts_len):
    rand_num = np.random.randint(60, 100)
    ts.append(rand_num)
```

```python
def gen_shapelet(ts, start_idx, shp_len=10):
    shp = ts[start_idx: start_idx+shp_len]
    for i in range(len(shp)):
        shp[i] += np.random.randn() * 5
    return shp
```

```python
#plot curve
x = list(range(len(ts)))
x_new = np.arange(0,ts_len-1, 0.1)
func = interpolate.interp1d(x,ts,kind='cubic')
ts_new = func(x_new)
x_new = list(range(len(ts_new)))
#plot ts
plt.plot(x_new, ts_new, color='lightgray')
plt.axis([-10, len(x_new)+10, 40, 115])

plt.savefig('structure.png', dpi=500)
```
